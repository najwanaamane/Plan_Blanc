# Position du problème

Un individu réalisable du problème est sous forme de matrice NxH où H désigne le nombre de chirurgiens et N le nombre de patients
A la position (i,j) de la matrice correspond la date du début de l'opération du patient i par le chirurgien j. Cette date est différent de 0 si et seulement si le chirurgien j a opéré le patient i.      

De plus étant donné qu'un patient ne peut pas se faire opérer par deux chirurgiens à la fois, alors nécessairement sur chaque ligne de la matrice, on aura H-1 valeurs nulles et une seule valeur non nulle.      

De meme dans une colonne, si au moins deux valeurs non nulles existent, alors la première en partant du haut doit etre supérieure à la seconde (c'est-à-dire que la date de l'operation du patient suivant doit etre superieure à la somme de la date de l'operation de son predecesseur plus la durée d'operation)      

Par ailleurs la date de l'operation du patient i par le chirurien j doit etre superieure à la fois à la date d'arrivée du patient et à la date d'arrivée du chirurgien à l'hopital tout en étant inférieur à sa date au plus tard.      

Et pour finir, la somme des durées des operations faites par un chirurgien doit etre inferieure à sa charge maximale de travail     

On définit 6 principales fonctions :   
    une fonction qui crée un individu    
    une fonction qui crée une population    
    une fonction qui trie chaque patient selon sa date au plus tard (les patients qui ont une date au plus tard petite seront priviligés car leurs cas sont plus graves)     
    une fonction qui permet juste de s'assurer que la population crée respecte les contraintes   
    une fonction qui evalue la performance des individus d'une population    
    une fonction qui determine le front de Pareto d'une population d'individus     


Notre plan blanc vise à optimiser un ensemble de 3 objectifs que sont :     
    1)-minimiser le nombre total de chirurgiens necessaires pour opérer tous les patients       
    2)-minimiser la somme des ecarts entre la date d'operation et la date d'arrivée d'un patient               
    

La performance d'un individu dans la population sera donc evaluée à partir des deux parametres definis ci-dessus     

#  Génerer un individu, une population- Fonctions Objectifs

    Un individu est un plan blanc réalisable

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Fonction qui génère un individu réalisable s'il en existe - N represente le nombre de patients et H le nombre de chirurgiens
def Init_Individu_White_Plan(N, H, early_date_patients, late_date_patients, date_surgeon, duration, load_charge):
    
    # Un individu est une matrice où le nombre de lignes== nombre de patients; nombre de colonnes== nombre de chirurgien
    # Chaque case represente la date à laquelle le patient a été opéré par un chirurgien donné
    Individu = np.zeros((N,H))
    # Initialisation des differents chirurgiens
    Bloc = [j for j in range(H)]
    Max_Load = [date_surgeon[h]+load_charge[h] for h in range(H)]
    for p in range(N):
        # Le premier patient à se faire opérer est souvent celui dont la date au plus tard est petite
        # Au début aucun chirurgien n'a encore fait d'opération
        if(p == 0):
            # On choisit pour ce patient le premier chirurgien arrivé (ce qui est logiquement normal)
            choice_chirurgien = date_surgeon.index(min(date_surgeon))
                
            # On choisit une date d'operation du patient 0 comprise dans sa fenetre de temps et superieure à la date d'arrivée du chirurgien
            a = max(date_surgeon[choice_chirurgien], early_date_patients[p])
            Q = a + duration[p] - Max_Load[choice_chirurgien] <= 0
            # Ici on s'assure que la date d'operation du patient ne dépasse pas sa date au plus tard
            if ( a > late_date_patients[p] or Q == False ):
                return "Failure of white plan!"
            else :
                b = min(Max_Load[choice_chirurgien]-duration[p], late_date_patients[p])
                d = list(np.linspace(a, b, 50000))
                Individu[p, choice_chirurgien] = int((d[1] + a)/2)
                    
        # Sinon on verifie si le chirurgien choisi n'a pas dejà effectué une operation
        else :
            # On choisit de facon aléatoire un chirurgien qui va operer le patient actuel
            choice_chirurgien = random.choices(Bloc, k = 1)[0]
            ind = Bloc.index(choice_chirurgien)
                
            # Ensemble des patients (eventuels ou non ) traités par le chirurgien avant le patient actuel
            previous_patients = list(Individu[:p, choice_chirurgien])
            
            if(max(previous_patients) == 0): # le chirurgien n'a encore traité aucun patient
                a = max(date_surgeon[choice_chirurgien], early_date_patients[p])
            else:
                treated = previous_patients.index(max(previous_patients))
                duree = Individu[treated, choice_chirurgien] + duration[treated]
                a = max(duree, early_date_patients[p])
                    
            count = 0
            Q = a + duration[p] - Max_Load[choice_chirurgien] <= 0
            while ( a > late_date_patients[p] or Q == False):
                count += 1
                del(Bloc[ind])
                # On verifie s'il existe un chirurgien qui peut operer le patient actuel sans dépasser sa date au plus tard avec la condition
                # que la charge maximale du chirurgien ne peut etre dépassé
                if (count >= H):
                    return "Failure of white plan. All patients can't be treated."
                        
                choice_chirurgien = random.choices(Bloc, k = 1)[0]
                ind = Bloc.index(choice_chirurgien)
                    
                # Ensemble des patients (eventuels ou non ) traités par le chirurgien avant le patient actuel
                previous_patients = list(Individu[:p, choice_chirurgien])
                    
                if(max(previous_patients) == 0):
                    a = max(date_surgeon[choice_chirurgien], early_date_patients[p])
                else:
                    treated = previous_patients.index(max(previous_patients))
                    duree = Individu[treated, choice_chirurgien] + duration[treated]
                    a = max(duree, early_date_patients[p])
                Q = a + duration[p] - Max_Load[choice_chirurgien] <= 0
                
            b = min(Max_Load[choice_chirurgien]-duration[p], late_date_patients[p])
            d = list(np.linspace(a, b, 50000))
            Individu[p, choice_chirurgien] = int((d[1]+a)/2)
                
        Bloc = [h for h in range(H)]
        
    return Individu

# Fonction qui initialise une population-Nbr represente la taille de la population
# Dans la fonction on a generé une population de taille le carré de celle spécifié
# En effet, si le plan blanc admet de solution, il peut arriver que la fonction qui genere un individu donne de solutions non realisables
    # car les dates d'operations sont choisies de facon aleatoire. Dans ce cas, la fonction ci-dessus va renvoyer en sortie :
    # "Aucune solution. Tous les patients ne peuvent pas etre traités. Failure of white plan."
# On impose alors que lors de la création de la population, la taille est élevée au carré pour obtenir une grande probabilté d'avoir effectivement 
    # la taille souhaitée d'individus realisables.
# Toujours est-il que on peut etre certain que si la population est vide, alors le plan blanc n'admet effectivement aucune solution
def Init_Population_White_Plan(Nbr, N, H, early_date_patients, late_date_patients, date_surgeon, duration, load_charge):
    
    Population, Generate, subset = [], [], []
    while(len(Generate) != Nbr*Nbr):
        Individu = Init_Individu_White_Plan(N, H, early_date_patients, late_date_patients, date_surgeon, duration, load_charge)
        Generate.append(Individu)
        
    for p in range(Nbr*Nbr):
        if( isinstance(Generate[p], str) != True): subset.append(Generate[p])
            
    if (len(subset) == 0): return 'Failure of white plan!'
    else:
        Population = subset[:Nbr]
        while(len(Population) < Nbr):
            Generate, subset = [], []
            while(len(Generate) != Nbr*Nbr):
                Individu = Init_Individu_White_Plan(N, H, early_date_patients, late_date_patients, date_surgeon, duration, load_charge)
                Generate.append(Individu)
            for p in range(Nbr*Nbr):
                if( isinstance(Generate[p], str) != True): subset.append(Generate[p])
            Population.extend(subset)
            
        return Population[:Nbr]

# Fonction qui trie les patients selon leur date au plus tard
def Order_passage_privilege(N, early_date_patients, late_date_patients, duration):
    tri_late_date_patients = []
    tri_early_date_patients = []
    tri_duration = []

    Sequence = []
    S = sum(late_date_patients)
    for j in range(N):
        tri_late_date_patients.append(min(late_date_patients))
        ind = late_date_patients.index(tri_late_date_patients[-1])

        tri_early_date_patients.append(early_date_patients[ind])
        tri_duration.append(duration[ind])
        late_date_patients[ind] = S
        Sequence.append(ind)
        
    return tri_early_date_patients, tri_late_date_patients, tri_duration, Sequence

# Fonction qui verifie que la population d'individus crees est realisable (respecte toutes les contraintes)
def Check_out_Population(Population, early_date_patients, late_date_patients, date_surgeon, duration, load_charge):
    Nbr = len(Population)
    N, H = Population[0].shape[0], Population[0].shape[1]
    final_truth = []
    Max_Load = [date_surgeon[h]+load_charge[h] for h in range(H)]
    # La fonction d'initialisation d'un individu assure que chaque patient n'est traité que par un seul chirurgien
    # Pas besoin de verification à ce niveau
    for s in range(Nbr):
        Individu = Population[s]
        
        truth_value = []
        for k in range(H):
            Dates = list(Individu[:,k])
            if (max(Dates) == 0):
                # Le chirurgien n'a traité aucun patient-sa charge maximale de travail n'est pas excédé
                truth_value.append(True)
            else :
                # Ensemble des patients traités par le chirurgien
                treated = [j for j in range(N) if Dates[j] != 0]
                T = len(treated)
                if (T == 1):
                    z = treated[0]
                    fenetre = (early_date_patients[z] <= Dates[z] and Dates[z] <= late_date_patients[z]) and (Dates[z] >= date_surgeon[k])
                    rest = Dates[z] + duration[z] <= Max_Load[k]
                    truth_value.append(fenetre and rest)
                else:
                    # Respect de la fenetre de temps
                    Window = [(early_date_patients[z] <= Dates[z] and Dates[z] <= late_date_patients[z]) and (Dates[z] >= date_surgeon[k])
                                                                                                              for z in treated]
                    # Respect de la succession des operations
                    Succession = [ Dates[treated[z+1]] >= Dates[treated[z]] + duration[treated[z]] for z in range(T-1)]
                    
                    # Respect de la charge maximale du chirurgien
                    rest = [Dates[treated[-1]] + duration[treated[-1]] <= Max_Load[k]]
                    Window, Succession, rest = np.array(Window), np.array(Succession), np.array(rest)
                    truth_value.append(Window.all() and Succession.all() and rest.all())
                    
        final_truth.append(np.array(truth_value).all())
    return np.array(final_truth).all()

# Fonction qui evalue la performance des individus de la population
def performance_population(Population, early_date_patient, Duration, Load_charge):
    Nbr, N, H = len(Population), Population[0].shape[0], Population[0].shape[1]
    # Un individu de la population a deux critères de performance (voire ci-dessus)
    Performance = np.zeros((Nbr,2))
    for s in range(Nbr):
        Individu = Population[s]
        # On determine le nombre de chirurgiens nécessaires pour opérer tous les patients
        number_chr = [ k for k in range(H) if max(list(Individu[:,k])) != 0 ]
        Performance[s,0] = len(number_chr)
        
        # On détermine la somme des ecarts entre les date d'operation et les dates d'arrivée des chirurgiens
        ecart_date = []
        for k in number_chr:
            patients = list(Individu[:,k])
            treated = [j for j in range(N) if patients[j] != 0]
            ecart = [ patients[z] - early_date_patient[z] for z in treated ]
            ecart_date.append(sum(ecart))
            
        Performance[s,1] = sum(ecart_date)
        
    return Performance

# Fonction qui trace le nuage des points en relation avec la performance des individus de la population
def nuages_de_performance(Population, early_date_patient, Duration, Load_charge):
    Nbr, N, H = len(Population), Population[0].shape[0], Population[0].shape[1]
    Performance = performance_population(Population, early_date_patient, Duration, Load_charge)
    
    plt.figure(figsize = (10,8))
    for s in range(Nbr):
        plt.scatter(Performance[s,0], Performance[s,1], marker = 'o')
        
    plt.xlabel('Nombre chirurgiens')
    plt.ylabel('Somme Ecarts dates operations')
    plt.show()

# Fonction qui determine le front de Pareto de la population
def Front_Pareto(Population, early_date_patient, Duration, Load_charge, draw):
    Performance = performance_population(Population, early_date_patient, Duration, Load_charge)
    Nbr = Performance.shape[0]
    Non_dominated = []
    for s in range(Nbr):
        perform = list(Performance[s,:])
        privated = list(Performance)
        del(privated[s])
        privated = [list(h) for h in privated]
        check = [ ((J[0]- perform[0] <= 0) and (J[1]- perform[1] < 0)) or ((J[0]- perform[0] < 0) and (J[1]- perform[1] <= 0)) for J in privated]
        if(np.array(check).any() != True):
            Non_dominated.append(s)
        perform, privated = [], []

    if (draw == 1):
        if(len(Non_dominated) == 0):
            plt.figure(figsize = (10,8))
            plt.scatter(list(Performance[:,0]), list(Performance[:,1]), marker = 'o', color = 'orange')
            plt.xlabel('Number of surgeons')
            plt.ylabel('Sum difference of dates')
            plt.title('Espace des fonctions objectifs')
            plt.show()
            print('Il a dominance mutuelle entre les solutions.')
        else:
            plt.figure(figsize = (10,8))
    
            plt.subplot(121)
            plt.scatter(list(Performance[:,0]), list(Performance[:,1]), marker = 'o', color = 'blue')
            plt.xlabel('Number of surgeons')
            plt.ylabel('Sum difference of dates')
            plt.title('Espace des fonctions objectifs')
            
            plt.subplot(122)
            image = [list(Performance[z,:]) for z in Non_dominated]
            none = [list(Performance[z,:]) for z in range(Nbr) if z not in Non_dominated]
            if(len(none) != 0):
                none, image = np.array(none), np.array(image)
                plt.scatter(list(none[:,0]), list(none[:,1]), marker = 'o', color = 'blue', label = 'Dominated_solutions')
                plt.plot(image[:,0], image[:,1], marker = 'o', color = 'red', label = 'Non_Dominated_solutions')
                plt.xlabel('Number of surgeons')
                plt.ylabel('Sum difference of dates')
                plt.title('Front de Pareto')
            else:
                image = np.array(image)
                plt.scatter(list(image[:,0]), list(image[:,1]), color = 'red')
                plt.plot(image[:,0], image[:,1], label = 'Non_Dominated_solutions-Pareto')
                plt.xlabel('Number of surgeons')
                plt.ylabel('Sum difference of dates')
                plt.title('Front de Pareto')
                print('Toutes les solutions sont non-dominées.')
    
            plt.legend()
            plt.tight_layout()
            plt.show()
    return Non_dominated
```

# Implementation de l'algorithme NSGA II pour le Plan Blanc

Pour l'algorithme, il reste à definir 5 fonctions principales que sont :       
    * La fonction de croisement entre deux individus selon une probabilté de croisement prédefinie      
    * La fonction de mutation d'un individu selon une probabilté de mutation prédefinie      
    * Une fonction qui rend realisable les enfants issus du croisement entre deux individus si ces derniers ne l'etaient pas     
    * Une fonction qui rend realisable l'individu issu de la mutation d'un individu si ce dernier ne l'etait pas     
    * Une fonction qui evalue la diversité de la population (en fonction de leur performance) en vue de la selection

En ce qui concerne la fonction qui evalue la diversité de la population, on peut soit utiliser la fonction usuelle dite de "crowding distance"  ou bien elaborer une autre fonction à implémenter.     

Par ailleurs on utilise le principe de selection aléatoire pour selectionner les individus qui doivent etre croisés entre eux. Autrement tous les individus de la population ont la meme chance d'etre pris pour un croisement. Il en est de meme pour la mutation.

```python
# fonction de croisement
    # A & B sont les individus à croiser- p represente la probabilté au dessus de laquelle A & B peuvent etre croisés
    # Le principe du croisement consiste à changer deux choses :
        # pour un ensemble de chirurgiens on modifie les patients qu'ils doivent operer
        # pour un ensemble de patients on modifie (ajout ou reduction) les chirurgiens qui doivent les operer ainsi que leurs dates d'operations
    # Il peut avoir des enfants non realisables après le croisement
    # Le croisement se fera à la moitié de la ligne et de la colonne (points de croisement : k,q = N//2, H//2)

def crossing(A, B, p):
    N, H = A.shape[0], A.shape[1]
    a, b = N//2, H//2
    child1, child2 = np.zeros((N, H)), np.zeros((N, H))
    aleatoire = random.uniform(0,1)
    if(aleatoire >= p):
        child1[:a, :b], child1[a:, b:] = A[:a, :b].copy(), A[a:, b:].copy()
        child1[:a, b:], child1[a:, :b] = B[:a, b:].copy(), B[a:, :b].copy()
        
        child2[:a, :b], child2[a:, b:] = B[:a, :b].copy(), B[a:, b:].copy()
        child2[:a, b:], child2[a:, :b] = A[:a, b:].copy(), A[a:, :b].copy()
    return A, B, child1, child2


def heuristique_crossing(A, early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge):
    N, H = A.shape[0], A.shape[1]
    Check = Check_out_Population([A], early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge)
    No_operated = [j for j in range(N) if max(list(A[j,:])) == 0]
    Bloc = [h for h in range(H)]
    if(Check == False or len(No_operated) != 0):
        Max_Load = [date_surgeon[h]+Load_charge[h] for h in range(H)]
        correct = A.copy()
        for p in range(N):
            Ht = [h for h in range(H) if correct[p,h] != 0]
            if(p == 0):
                if(len(Ht) != 0):
                    dates = [correct[0,h] for h in Ht]
                    ind = dates.index(min(dates))
                    least = Ht[ind]
                    privated = [h for h in Ht if h != least]
                    for h in privated : correct[0,h] = 0
                else:
                    first = date_surgeon.index(min(date_surgeon))
                    a = max(date_surgeon[first], early_date_patient[0])
                    Q = a + duration[0] - Max_Load[first] <= 0
                    if(a <= late_date_patients[0] and Q == True):
                        b = min(Max_Load[first] - Duration[p], late_date_patients[0])
                        Op = ramdom.uniform(a, b)
                        d = list(np.linspace(a, Op, 5000))
                        correct[0,first] = int((d[50]+a)/2)
                    else : return None
            else:
                if(len(Ht) != 0):
                    dates = [correct[p,h] for h in Ht]
                    ind = dates.index(min(dates))
                    least = Ht[ind]
                    privated = [h for h in Ht if h != least]
                    for h in privated : correct[p,h] = 0
                    a, Rest = 0, 0
                    if(np.max(correct[:p, least]) == 0):
                        a = max(date_surgeon[least], early_date_patient[p])
                        Rest = a + Duration[p] - Max_Load[least] <= 0
                    else:
                        last = list(correct[:p, least]).index(max(list(correct[:p, least])))
                        Rest = correct[last,least] + Duration[last] + Duration[p] <= Max_Load[least]
                        a = max(correct[last,least] + Duration[last], early_date_patient[p])
                    count = 0
                    while(a > late_date_patients[p] or Rest == False):
                        count += 1
                        if(count >= H):
                            return None
                        else:
                            Bloc.remove(least)
                            least = random.choices(Bloc, k = 1)[0]
                            if(np.max(correct[:p, least]) == 0):
                                a = max(date_surgeon[least], early_date_patient[p])
                                Rest = a + Duration[p] - Max_Load[least] <= 0
                            else:
                                last = list(correct[:p,least]).index(max(list(correct[:p, least])))
                                Rest = correct[last,least] + Duration[last] + Duration[p] <= Max_Load[least]
                                a = max(correct[last,least] + Duration[last], early_date_patient[p])
                                    
                    b = min(Max_Load[least] - Duration[p], late_date_patients[p])
                    d = list(np.linspace(a, b, 5000))
                    correct[p,least] = int(d[50])
                    Bloc = [h for h in range(H)]
                else:
                    a, Rest = 0, 0
                    select = random.choices(Bloc, k = 1)[0]
                    if(np.max(correct[:p, select]) == 0):
                        a = max(date_surgeon[select], early_date_patient[p])
                        Rest = a + Duration[p] - Max_Load[select] <= 0
                    else:
                        last = list(correct[:p,select]).index(max(list(correct[:p, select])))
                        Rest = correct[last,select] + Duration[last] + Duration[p] <= Max_Load[select]
                        a = max(correct[last,select] + Duration[last], early_date_patient[p])
                    count = 0
                    while(a > late_date_patients[p] or Rest == False):
                        count += 1
                        if(count >= H): return None
                        else:
                            Bloc.remove(select)
                            select = random.choices(Bloc, k = 1)[0]
                            if(np.max(correct[:p, select]) == 0):
                                a = max(date_surgeon[select], early_date_patient[p])
                                Rest = a + Duration[p] - Max_Load[select] <= 0
                            else:
                                last = list(correct[:p,select]).index(max(list(correct[:p, select])))
                                Rest = correct[last,select] + Duration[last] + Duration[p] <= Max_Load[select]
                                a = max(correct[last,select] + Duration[last], early_date_patient[p])
                                
                    b = min(Max_Load[select] - Duration[p], late_date_patients[p])
                    d = list(np.linspace(a, b, 5000))
                    correct[p,select] = int(d[50])
                    Bloc = [h for h in range(H)]
                    
        V = Check_out_Population([correct], early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge)
        if(V == True): return correct
        else : return None
    else: return A
        

# fonction de mutation
    # A represente l'individu qui doit se faire muter - q la probabilté au dessus de laquelle A peut etre muté
    # Le principe de la mutation vise à échanger entre deux chirurgiens les patients opérés
    # Cette methode nous donne une grande chance que l'individu muté soit réalisable

def mutate(A, q):
    N, H = A.shape[0], A.shape[1]
    a = N//2
    mute = np.zeros((N, H))

    set_surgeon = [j for j in range(H)]
    # Choix des deux chirurgiens qui interviendront dans la mutation
    h1 = random.choices(set_surgeon, k = 1)[0]
    # On s'assure que le meme chirurgien ne sera pas encore choisi
    del(set_surgeon[h1])
    h2 = random.choices(set_surgeon, k = 1)[0]
    aleatoire = random.uniform(0,1)
    if(aleatoire >= q):
        mute = A.copy()
        mute[a:,h1], mute[a:,h2] = A[a:,h2], A[a:,h1]
    return A, mute, h1, h2


# heuristique pour rendre realisable l'individu issu de la mutation

def heuristique_mutate(A, h1, h2, early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge):
    N, H = A.shape[0], A.shape[1]
    L, c = [h1, h2], N//2
    Check = Check_out_Population([A], early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge)
    Bloc = [k for k in range(H)]
    Max_Load = [date_surgeon[h]+Load_charge[h] for h in range(H)]
    if(Check == False):
        correct = A.copy()
        for h in L:
            Ht = [j for j in range(c, N) if correct[j, h] != 0]
            if(len(Ht) == 0) : continue
            else:
                a, Rest = 0, 0
                for z in Ht:
                    if(np.max(correct[:z, h]) == 0):
                        a = max(date_surgeon[h], early_date_patient[z])
                        Rest = a + Duration[z] - Max_Load[h] <= 0
                    else:
                        last = list(correct[:z, h]).index(max(list(correct[:z, h])))
                        Rest = correct[last,h] + Duration[last] + Duration[z] <= Max_Load[h]
                        a = max(correct[last,h] + Duration[last], early_date_patient[z])
                    count, least = 0, h
                    while(a > late_date_patients[z] or Rest == False):
                        count += 1
                        if(count >= H): return None
                        else:
                            Bloc.remove(least)
                            least = random.choices(Bloc, k = 1)[0]
                            if(np.max(correct[:z, least]) == 0):
                                a = max(date_surgeon[least], early_date_patient[z])
                                Rest = a + Duration[z] - Max_Load[least] <= 0
                            else:
                                last = list(correct[:z,least]).index(max(list(correct[:z, least])))
                                Rest = correct[last,least] + Duration[last] + Duration[z] <= Max_Load[least]
                                a = max(correct[last,least] + Duration[last], early_date_patient[z])
                                
                    b = min(Max_Load[least] - Duration[z], late_date_patients[z])
                    d = list(np.linspace(a, b, 5000))
                    correct[z,least] = int(d[50])
                    Bloc = [q for q in range(H)]
                    
        V = Check_out_Population([correct], early_date_patient, late_date_patients, date_surgeon, Duration, Load_charge)
        if(V == True): return correct
        else : return None
    else:
        return A



# fonction qui divise la population en plusieurs fronts de non_domination

def divide_non_dominated(Population, early_date_patient, Duration, Load_charge, draw):

    Levels_Non_dominated, Levels_Performance = [], []
    
    Performance_all, N = performance_population(Population, early_date_patient, Duration, Load_charge), len(Population)
    Non_dominated = Front_Pareto(Population, early_date_patient, Duration, Load_charge, draw)
    dominated = [j for j in range(N) if j not in Non_dominated]
    
    best_subset, least_subset = [Population[j] for j in Non_dominated], [Population[j] for j in dominated]
    perform_non_dominated = [list(Performance_all[j,:]) for j in Non_dominated]
    perform_non_dominated = np.array(perform_non_dominated)
    
    Levels_Non_dominated.append(best_subset), Levels_Performance.append(perform_non_dominated)
    
    while(len(least_subset) != 0):
        Performance_all, N = performance_population(least_subset, early_date_patient, Duration, Load_charge), len(least_subset)
        Non_dominated = Front_Pareto(least_subset, early_date_patient, Duration, Load_charge, draw)
        dominated = [j for j in range(N) if j not in Non_dominated]
        
        best_subset, least_subset = [least_subset[j] for j in Non_dominated], [least_subset[j] for j in dominated]
        perform_non_dominated = [list(Performance_all[j,:]) for j in Non_dominated]
        perform_non_dominated = np.array(perform_non_dominated)
        
        Levels_Non_dominated.append(best_subset), Levels_Performance.append(perform_non_dominated)
        
    return Levels_Non_dominated, Levels_Performance



# fonction de diversité "crowding-distance" pour assurer que les individus sélectionnés sont diversifiés
    # Cela empeche de rester bolqué à un minimum local et d'explorer davantage l'espace des solutions

def crowding_distance(Levels_Non_dominated, Levels_Performance, arg):
    Performance = Levels_Performance[arg]
    Non_dominated = Levels_Non_dominated[arg]
    W = len(Non_dominated)
    
    # Initialisation des distances à 0
    distance = [0 for _ in range(W)]
    
    # Sequence des individus à prendre dans l'ordre de priorité
    Sequence_prior = []
    
    for k in range (2):
        # Valeurs de la fonction objectif k
        projection, copy = list(Performance[:,k]), list(Performance[:,k])
        # Tri des valeurs des individus
        projection.sort()
        # Individus triés dans l'ordre croissant selon la valeur de la performance k
        sequence_sort = []
        for i in range(W):
            # Prendre l'individu correspondant à la valeur projection[i]
            sequence_sort.append(copy.index(projection[i]))
            # S'assurer que l'individu n'est plus choisi une seconde fois
            copy[sequence_sort[-1]] = float('inf')
        # Attribuer une distance infinie au minimum et au maximum pour la performance k
        distance[sequence_sort[0]] += float('inf')
        distance[sequence_sort[W-1]] += float('inf')
        maxi, mini = max(projection), min(projection)
        diff = maxi - mini
        for j in range(1,W-1):
            if (diff != 0):distance[sequence_sort[j]] += (projection[j+1] - projection[j-1])/diff
            else : distance[sequence_sort[j]] += float('inf')
    for j in range(W):
        Sequence_prior.append(distance.index(max(distance)))
        distance[Sequence_prior[-1]] = -1
        
    return Sequence_prior
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def White_Plan(Tmax, Nbr, N, H, early_date_patients, late_date_patients, date_surgeon, duration, load_charge, pc, pm, draw):

    # Tmax : maximun d'iterations
    # N_selected : nombre d'individus a selectionner pour le croisement et la mutation - doit etre inferieur à Nbr
    # Nbr : nombre initial d'individus dans la population
    # Trier les patients selon leurs dates au plus tard
    EDP, LDP, Du, Sequence = Order_passage_privilege(N, early_date_patients, late_date_patients, duration)
    
    # Initialisation de la population
    POPULATION = Init_Population_White_Plan(Nbr, N, H, EDP, LDP, date_surgeon, Du, load_charge)
    if(isinstance(POPULATION, str) == True):
        return 'Failure of White Plan!', 'Void', ''
    else:
        t = 0
        Evolution, First, Second = ['Before Croosing_Mutation'], [], []
        while(t < Tmax):
            # Evaluation de la population  & division en plusieurs niveaux de non domination
            Levels_non_Dominated, Perform_Non_Dominated = divide_non_dominated(POPULATION, EDP, Du, load_charge, 0)
            First.append(np.min(Perform_Non_Dominated[0][:,0])), Second.append(np.min(Perform_Non_Dominated[0][:,1]))
            
            if (t == 0):
                Selected, T = [], len(Levels_non_Dominated)
                for j in range(T):
                    Selected.extend(Levels_non_Dominated[j])
            else: 
                Evolution.append(f'After {t} th Crossing_Mutation')
                # Selection des individus selon leur rangs de non  domination et en fonction de la distance de crowding
                Selected = []
                    # Si tous les individus sont non dominés ou si le premier niveau de non domination  >= Nbr
                if(len(Levels_non_Dominated) == 1 or len(Levels_non_Dominated[0]) >= Nbr): 
                    # Utiliser la distance de crowding pour les départager selon un ordre de priorité
                    Seq_prior = crowding_distance(Levels_non_Dominated, Perform_Non_Dominated, 0)
                    # Selection des individus à croiser et à muter
                    taken = [Levels_non_Dominated[0][Seq_prior[s]] for s in range(Nbr)]
                    Selected.extend(taken)
                else:
                    Selected.extend(Levels_non_Dominated[0])
                    diff = Nbr - len(Levels_non_Dominated[0])
                    j = 1
                    while(len(Selected) != Nbr):
                        if(len(Levels_non_Dominated[j]) >= diff):
                            Seq_prior = crowding_distance(Levels_non_Dominated, Perform_Non_Dominated, j)
                            taken = [Levels_non_Dominated[j][Seq_prior[s]] for s in range(diff)]
                        else:
                            T = len(Levels_non_Dominated[j])
                            Seq_prior = crowding_distance(Levels_non_Dominated, Perform_Non_Dominated, j)
                            taken = [Levels_non_Dominated[j][Seq_prior[s]] for s in range(T)]
                            
                        j += 1
                        Selected.extend(taken)
                        diff -= len(taken)
            
            # Une nouvelle population de taille 2*N sera crée(N parents et N enfants issus du croisement et de la mutation)
            # Mutation et croisement entre les individus - la moitié des individus seront croisés entre eux ; le reste muté
            q, r = divmod(Nbr, 2)
            # Les q+r premiers individus étant les meilleurs, ils seront croisés entre eux(selection de croisement par rang)
            Cross_Mute_population = []
            for A in Selected[:q+r]:
                privated = [B for B in Selected[:q+r] if (B-A == 0).all() != True]
                for B in privated:
                    _, _, c1, c2 = crossing(A, B, pc)
                    Cross = [c1, c2]
                    for c in Cross:
                        Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                        if(type(Cor) != type(None)) : Cross_Mute_population.append(Cor)
                        else:
                            for _ in range(50):
                                Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                                if(type(Cor) != type(None)) : 
                                    Cross_Mute_population.append(Cor)
                                    break
                                
            # Le reste des individus seront mutés
            for A in Selected[q+r:]:
                _, mute, h1, h2 = mutate(A, pm) 
                Mute = heuristique_mutate(mute, h1, h2, EDP, LDP, date_surgeon, Du, load_charge)
                
                if(type(Mute) != type(None)): Cross_Mute_population.append(Mute)
                else:
                    # On donne 50 chances à l'individu pour que sa mutation reussisse
                    # (la mutation fait intervenir deux chirurgiens d'un plan blanc de facon aleatoire)
                    for _ in range(50):
                        _, mute, h1, h2 = mutate(A, pm)
                        Mute = heuristique_mutate(mute, h1, h2, EDP, LDP, date_surgeon, Du, load_charge)
                        if(type(Mute) != type(None)):
                            Cross_Mute_population.append(Mute)
                            break
                        
            # Si la taille de la generation d'individus engendrés est inferieure strictement à N, on mute les q+r premiers individus
                # on croise d'abord les q restants - ensuite on croise les q+r meilleurs avec les q restants
            if(len(Cross_Mute_population) < Nbr):
                for A in Selected[:q+r]:
                    for B in Selected[q+r:]:
                        _, _, c1, c2 = crossing(A, B, pc)
                        Cross = [c1, c2]
                        for c in Cross:
                            Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                            if(type(Cor) != type(None)) : Cross_Mute_population.append(Cor)
                            else:
                                for _ in range(20):
                                    Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                                    if(type(Cor) != type(None)) : 
                                        Cross_Mute_population.append(Cor)
                                        break
                                        
                for A in Selected[:q+r]:
                    _, mute, h1, h2 = mutate(A, pm)
                    Mute = heuristique_mutate(mute, h1, h2, EDP, LDP, date_surgeon, Du, load_charge)
                    if(type(Mute) != type(None)): Cross_Mute_population.append(Mute)
                    else:
                        for _ in range(50):
                            _, mute, h1, h2 = mutate(A, pm)
                            Mute = heuristique_mutate(mute, h1, h2, EDP, LDP, date_surgeon, Du, load_charge)
                            if(type(Mute) != type(None)):
                                Cross_Mute_population.append(Mute)
                                break
                                
                for A in Selected[q+r:]:
                    privated = [B for B in Selected[q+r:] if (B-A == 0).all() != True]
                    for B in privated:
                        _, _, c1, c2 = crossing(A, B, pc)
                        Cross = [c1, c2]
                        for c in Cross:
                            Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                            if(type(Cor) != type(None)) : Cross_Mute_population.append(Cor)
                            else:
                                for _ in range(50):
                                    Cor = heuristique_crossing(c, EDP, LDP, date_surgeon, Du, load_charge)
                                    if(type(Cor) != type(None)) : 
                                        Cross_Mute_population.append(Cor)
                                        break
                                              
                                        
            POPULATION = []
            POPULATION.extend(Selected), POPULATION.extend(Cross_Mute_population[:Nbr])
                
            t += 1
            
        # Evaluation finale de la population  & division en plusieurs niveaux de non domination
        Levels_non_Dominated, Perform_Non_Dominated = divide_non_dominated(POPULATION, EDP, Du, load_charge, 0)
        First.append(np.min(Perform_Non_Dominated[0][:,0])), Second.append(np.min(Perform_Non_Dominated[0][:,1]))
        Evolution.append(f'After {t} th Crossing_Mutation')

        if (draw == 1):
            plt.figure(figsize = (20,10))
            plt.subplot(121)
            plt.plot([t for t in range(Tmax+1)], First, linestyle = '-', marker = 'o', color = 'blue')
            plt.ylabel('Values of first objective function')
            plt.xticks([t for t in range(Tmax+1)], labels = Evolution, rotation = 90)
            plt.title('Evolution of the fisrt objective')
    
            plt.subplot(122)
            plt.plot([t for t in range(Tmax+1)], Second, linestyle = '-', marker = 'o', color = 'blue')
            plt.ylabel('Values of second objective function')
            plt.xticks([t for t in range(Tmax+1)], labels = Evolution, rotation = 90)
            plt.title('Evolution of the second objective')
            
            print('\t','View of performance of best White Plans :')
            Id = [k+1 for k in range(len(Levels_non_Dominated[0]))]
            first = [one for one in list(np.round(Perform_Non_Dominated[0][:,0],2))]
            second = [second for second in list(np.round(Perform_Non_Dominated[0][:,1],2))]
            table = {'Id White Plans': Id, 'Number of Surgeons used': first, 'Delay of surgical interventions': second}
            
            Data = pd.DataFrame(table)
            
            print(Data.tail().to_string(index = False))
        
        # La meilleure population se trouve au premier niveau de non domination
        return Levels_non_Dominated[0], Perform_Non_Dominated[0], Sequence

def White_Plan_Visualizer(Levels_non_Dominated, Sequence, Ds, Ld, duration, arg, Name):
    # Ds - date d'arrivée des chirurgiens
    # Ld - charge de travail maximale
    if(arg > len(Levels_non_Dominated)): return f'Out of range!. The length of non dominated solutions is {len(Levels_non_Dominated)}.'
    else:
        Plan_blanc_selected = Levels_non_Dominated[arg-1]
        N, H = Plan_blanc_selected.shape[0], Plan_blanc_selected.shape[1]
        ML = [Ds[h]+Ld[h] for h in range(H)]
        Mapping = plt.cm.jet
        
        colors = Mapping(np.linspace(0, 1, H))
        
        total_set, time = [], []
        name = Name + '_' + str(arg) + '.pdf'
        with PdfPages(name) as pdf:
            for k in range(H):
                set_patients, Set_dates, end = [], [], []
                Dates = list(Plan_blanc_selected[:,k])
                if (max(Dates) == 0):
                    plt.figure(figsize = (20, 20))
                    plt.plot(0, 0, label = f"Surgeon {k+1}:\n Heure d'arrivee {Ds[k]} mn \n Maximum_charge {Ld[k]} mn\n Arret travail {ML[k]} mn",
                             color = colors[k])
                    time.append(str(0) + ' min')
                else:
                    ind_treated = [j for j in range(N) if Dates[j] != 0]
                    treated = [Sequence[j]+1 for j in ind_treated]
                    set_patients.extend(treated)
                    
                    Set_dates.extend([Dates[z] for z in ind_treated])
                    
                    end = [Set_dates[j] + duration[Sequence[ind_treated[j]]] for j in range(len(ind_treated))]
                    string = [str(j) for j in treated]
                    plt.figure(figsize = (20, 20))
                    for j in range(len(treated)):
                        plt.barh(string[j], width = (end[j]-Set_dates[j]), left = Set_dates[j], color = colors[k], height = 0.2)
                    D, E = [0], Set_dates.copy()
                    D.extend(E)
                    D.extend(end)
                    plt.xticks(np.array(D))
                    Labels = plt.gca().get_xticklabels()
                    Labels[0] = '\n Début\n Catastrophe:\n T = 0 mn'
                    
                    plt.gca().set_xticklabels(Labels)
                    plt.ylabel('Patients')
                    plt.xlabel('Succession operations since the beginning of desaster at T = 0 min')
                    plt.title(f"Surgeon's sequence operation {k+1}")
                    plt.tight_layout()
                    pdf.savefig()
                    plt.close()
                    plt.figure(figsize = (20, 20))
                    plt.plot(0, 0, label = f"Surgeon {k+1}, :\n Heure d'arrivee {Ds[k]} mn\n Maximum_charge {Ld[k]} mn\n Arret travail {ML[k]} mn", 
                             color = colors[k])
                    plt.plot(np.array(treated), np.array(Set_dates), marker = 'o', color = colors[k], linestyle = '-')
                    Time_computed = [duration[Sequence[j]] for j in ind_treated]
                    time.append(sum(Time_computed))
                total_set.append(len(set_patients))
                plt.xlabel("Patients")
                plt.ylabel("Operations dates of patients since the beginning of desaster at T = 0 min")
                plt.xticks(np.array((set_patients)), rotation = 90)
                plt.yticks(np.array(Set_dates))
                plt.title(f"Surgeon's sequence operation {k+1}")
                plt.legend()
                plt.tight_layout()
                pdf.savefig()
                plt.close()
                
            plt.figure(figsize = (20, 20))
            Bloc = [f'Surgeon {k+1}' for k in range(H)]
            bar = plt.bar([k+1 for k in range(H)], total_set, color = colors)
            plt.xlabel("Surgeons")
            plt.ylabel('Number of patients operated by each surgeon')
            plt.xticks([k+1 for k in range(H)])
            plt.title(f'Synthesis of White Plan {arg}')
            plt.legend(bar, Bloc)
            pdf.savefig()
            plt.close()
            
            plt.figure(figsize = (20, 20))
            Bloc = [f'Surgeon {k+1}' for k in range(H)]
            bar = plt.bar([k+1 for k in range(H)], time, color = colors)
            plt.xlabel("Surgeons")
            plt.ylabel('Total duration of operations by each surgeon in min')
            plt.xticks([k+1 for k in range(H)])
            plt.yticks(time)
            plt.title(f'Synthesis of White Plan {arg}')
            plt.legend(bar, Bloc)
            pdf.savefig()
            plt.close()
```
